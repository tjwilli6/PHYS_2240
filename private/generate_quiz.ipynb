```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("/Users/tjwilli/presentation.mplstyle")
```

```python
angle2vert1 = np.random.randint(25,65)
angle2vert2 = np.random.randint(25,65)
while np.abs(angle2vert2-angle2vert1) < 10:
    angle2vert2 = np.random.randint(25,65)
mass = np.random.randint(2,25)
speed1 = np.random.randint(10,30)
speed2 = np.random.randint(10,30)
while np.abs(speed1-speed2) < 5:
    speed2 = np.random.randint(10,30)
```

```python
def plot():
    #plt.figure(figsize=(8,12))
    theta_x1 = 90 - angle2vert1 + 180
    vx1 = np.cos(np.deg2rad(theta_x1))
    vy1 = np.sin(np.deg2rad(theta_x1))
    plt.arrow(0,0,vx1,vy1,head_width=0.05,facecolor='blue',edgecolor='blue')
    textboxangle1 = 90 - angle2vert1/2
    xs1 = (np.cos(np.deg2rad(textboxangle1))/4 + vx1)
    ys1 = (np.sin(np.deg2rad(textboxangle1))/4 + vy1)
    plt.text(xs1,ys1,r'$\theta_i$={}$^\circ$'.format(angle2vert1),ha='center')
    theta_x2 = angle2vert2 + 90
    vx2 = np.cos(np.deg2rad(theta_x2))
    vy2 = np.sin(np.deg2rad(theta_x2))
    plt.arrow(vx1,vy1,vx2,vy2,head_width=0.05,facecolor='green',edgecolor='green')
    textboxangle2 = (angle2vert2/2) + 90
    xs2 = (np.cos(np.deg2rad(textboxangle2))/4 + vx1)
    ys2 = (np.sin(np.deg2rad(textboxangle2))/4 + vy1)
    plt.text(xs2,ys2,r'$\theta_f=${}$^\circ$'.format(angle2vert2),ha='center')
    
    plt.xlim((vx1+vx2)*1.1,0.1)
    plt.ylim(vy1*1.1,(vy1+vy2*1.1))
    ymin = plt.ylim()[0]
    plt.gca().set_aspect('equal')
    plt.gca().xaxis.set_visible(False)
    plt.gca().yaxis.set_visible(False)
    plt.plot([vx1,vx1],[vy1,plt.ylim()[1]],c='k',ls='--')
    plt.text(.1,ymin+0.05,"initial speed (blue) = {} m/s".format(speed1),ha='right',color='blue')
    plt.text(.1,ymin+0.1,"final speed (green) = {} m/s".format(speed2),ha='right',color='green')
    plt.text(0.1,ymin+0.15,"mass = {} g".format(mass),ha='right')
def print_solution():
    theta_x1 = 90 - angle2vert1 + 180
    print('THETA_X1 = {} degrees'.format(theta_x1))
    vhx1,vhy1 = np.cos(np.deg2rad(theta_x1)),np.sin(np.deg2rad(theta_x1))
    print('vhat1 = <{:.3f},{:.3f}>'.format(vhx1,vhy1))
    vx1,vy1 = speed1 * vhx1, speed1 * vhy1
    print('v1 = <{:.3f},{:.3f}> m/s'.format(vx1,vy1))
    px1,py1 = mass * vx1 / 1000, mass * vy1 / 1000
    print('p1 = <{:.3f},{:.3f}> kg m/s'.format(px1,py1))
    print('-'*30)
    theta_x2 = angle2vert2 + 90
    print('THETA_X2 = {} degrees'.format(theta_x2))
    vhx2,vhy2 = np.cos(np.deg2rad(theta_x2)),np.sin(np.deg2rad(theta_x2))
    print('vhat2 = <{:.3f},{:.3f}>'.format(vhx2,vhy2))
    vx2,vy2 = speed2 * vhx2, speed2 * vhy2
    print('v2 = <{:.3f},{:.3f}> m/s'.format(vx2,vy2))
    px2,py2 = mass * vx2 / 1000, mass * vy2 / 1000
    print('p2 = <{:.3f},{:.3f}> kg m/s'.format(px2,py2))
    print('-'*30)
    print('-'*30)
    print('DeltaP = <{:.3f},{:.3f}> kg m/s'.format(px2-px1,py2-py1))
```

```python
plot()
```

```python
print_solution()
```
